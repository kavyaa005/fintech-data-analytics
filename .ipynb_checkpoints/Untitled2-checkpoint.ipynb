```python
import pandas as pd
```

```python
nav = pd.read_csv("Data/raw/02_nav_history.csv")

print(nav.head())
```

```python
print(nav.shape)
print(nav.info())
```

```python
nav.isnull().sum()
```

```python
nav['date'] = pd.to_datetime(
    nav['date'],
    errors='coerce'
)

print(nav.dtypes)
```

```python
print("Rows before:", len(nav))

duplicates = nav.duplicated(
    subset=['amfi_code', 'date']
).sum()

print("Duplicate rows:", duplicates)
```

```python
print("Minimum NAV:", nav['nav'].min())
print("Maximum NAV:", nav['nav'].max())
```

```python
import os

os.makedirs("Data/processed", exist_ok=True)

nav.to_csv(
    "Data/processed/nav_history_clean.csv",
    index=False
)

print("NAV file cleaned and saved successfully")
```

```python
perf = pd.read_csv(
    "Data/raw/07_scheme_performance.csv"
)

print(perf.shape)
print(perf.head())
```

```python
perf.isnull().sum()
```

```python
txn['transaction_type'].value_counts()
```

```python
txn['kyc_status'].value_counts()
```

```python
perf.isnull().sum()
```

```python
txn.to_csv(
    "Data/processed/investor_transactions_clean.csv",
    index=False
)
```

```python
return_cols = [
    'return_1yr_pct',
    'return_3yr_pct',
    'return_5yr_pct'
]

for col in return_cols:
    perf[col] = pd.to_numeric(
        perf[col],
        errors='coerce'
    )

print(
    perf[return_cols].dtypes
)
```

```python
anomalies = perf[
    (perf['expense_ratio_pct'] < 0.1)
    |
    (perf['expense_ratio_pct'] > 2.5)
]

print(anomalies.shape)
anomalies
```

```python
perf['expense_flag'] = (
    (perf['expense_ratio_pct'] < 0.1)
    |
    (perf['expense_ratio_pct'] > 2.5)
)
```

```python
perf.to_csv(
    "Data/processed/scheme_performance_clean.csv",
    index=False
)

print("Performance file saved")
```
