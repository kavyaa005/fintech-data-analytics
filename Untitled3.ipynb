```python
from sqlalchemy import create_engine

engine = create_engine(
    "sqlite:///bluestock_mf.db"
)

print("Database Created Successfully")
```

```python
import pandas as pd

nav = pd.read_csv(
    "Data/processed/nav_history_clean.csv"
)

nav.to_sql(
    "fact_nav",
    engine,
    if_exists="replace",
    index=False
)

print("NAV table loaded")
```

```python
count = pd.read_sql(
    "SELECT COUNT(*) as rows FROM fact_nav",
    engine
)

print(count)
```

```python
fund = pd.read_csv("Data/raw/01_fund_master.csv")
fund.to_sql("dim_fund", engine, if_exists="replace", index=False)

txn = pd.read_csv("Data/processed/investor_transactions_clean.csv")
txn.to_sql("fact_transactions", engine, if_exists="replace", index=False)

perf = pd.read_csv("Data/processed/scheme_performance_clean.csv")
perf.to_sql("fact_performance", engine, if_exists="replace", index=False)

print("All tables loaded")
```

```python
for table in [
    "dim_fund",
    "fact_nav",
    "fact_transactions",
    "fact_performance"
]:
    count = pd.read_sql(
        f"SELECT COUNT(*) as cnt FROM {table}",
        engine
    )

    print(table)
    print(count)
    print("-" * 30)
```
